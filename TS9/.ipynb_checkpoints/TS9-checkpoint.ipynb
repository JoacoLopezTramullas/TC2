```python
from scipy import signal
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

# Agrego funciones que voy a usar de splane
from splane import plot_plantilla

# Ajusto tamaño de graficos
mpl.rcParams['figure.figsize'] = (13,13)
mpl.rcParams['font.size'] = 15
```

## Filtro IIR

```python
fs = 1000
f_pass = [2.5, 26.5]
f_stop = [1, 35]
gpass = 1
gstop = 40

orden, wn = signal.buttord(f_pass, f_stop, gpass, gstop, fs = fs)
print("Ordendel filtro:"  ,orden)

filtro = signal.butter(orden, wn, 'band', analog = False, output='sos', fs = fs)

w_iir,h_iir = signal.sosfreqz(filtro, fs=fs)

#plot_plantilla('bandpass', f_pass, 0.5, f_stop, 40, fs = 1000)

# Ploteo
plt.close('all')

# Imprimo plantilla
plt.fill([0, 1, 1,35,35,100,100,25,25,3,3,0],[-40,-40,5,5,-40,-40,-60,-60,0,0,-60,-60],color=[0.8,0.902,0.8])

## Imprimo IIR 
plt.title('Respuesta de módulo')
plt.plot(w_iir, 20 * np.log10(abs(h_iir)))
plt.ylabel('Amplitud [dB]')
plt.xlabel('Frecuencia [Hz]')
plt.legend(['Bw digital', 'Butter IIR orden 17'])
plt.xlim([0,100])
plt.ylim([-60,5])
plt.grid(True)

```

## Filtro FIR

```python
from scipy import signal
import matplotlib.pyplot as plt

# Ajusto tamaño de graficos
mpl.rcParams['figure.figsize'] = (13,13)
mpl.rcParams['font.size'] = 15

fs = 1000.0  # Hz
taps = 1701
bands = (0, 1, 2.5, 25, 27, 500)
gains = (0, 0.01, 1, 1, 0, 0)

num_z = signal.firls(taps, bands, gains, fs=fs)
den_z = 1

# Obtengo H(z)
w_fir , h_fir = signal.freqz(num_z, den_z, fs=fs)

# Ploteo
plt.close('all')

# Imprimo plantilla
plt.fill([0, 1, 1,35,35,100,100,25,25,3,3,0],[-40,-40,5,5,-40,-40,-60,-60,0,0,-60,-60],color=[0.8,0.902,0.8])

# Imprimo FIR
plt.title('Respuesta de módulo')
plt.plot(w_fir, 20 * np.log10(abs(h_fir)), 'b')
plt.ylabel('Amplitud [dB]')
plt.xlabel('Frecuencia [Hz]')
plt.legend(['Bw digital', 'FIR orden 1701'])
plt.xlim([0,100])
plt.ylim([-60,5])
plt.grid(True)
```

## IIR y FIR

```python
# Ploteo
plt.close('all')

# Imprimo plantilla
plt.fill([0, 1, 1,35,35,100,100,25,25,3,3,0],[-40,-40,5,5,-40,-40,-60,-60,0,0,-60,-60],color=[0.8,0.902,0.8])

## Imprimo IIR  y FIR
plt.title('Respuesta de módulo')
plt.plot(w_iir, 20 * np.log10(abs(h_iir)), 'r')
plt.plot(w_fir, 20 * np.log10(abs(h_fir)), 'b')
plt.ylabel('Amplitud [dB]')
plt.xlabel('Frecuencia [Hz]')
plt.legend(['Bw digital', 'Butter IIR orden 17', 'FIR orden 1701'])
plt.xlim([0,100])
plt.ylim([-60,5])
plt.grid(True)
```
