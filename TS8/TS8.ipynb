```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt

# módulo de SciPy
from scipy import signal as sig

# Agrego funciones que voy a usar de splane
from splane import bodePlot, analyze_sys

# Ajusto tamaño de graficos
mpl.rcParams['figure.figsize'] = (13,13)
mpl.rcParams['font.size'] = 15
```

### Ejercicio 2



#### a) fs = 100 KHZ

```python
# Defino parámetros
fs = 1e5
wo = 2*np.pi*1000
Q = 1/np.sqrt(2)

# Defino num y den de H(s)
num_s = np.array([wo**2])
den_s = np.array([1,wo/Q,wo**2])

# Obtengo H(s)
w_s , h_s = sig.freqs(num_s, den_s, worN=np.logspace(-1, 7, 1000))

# Obtengo num y den de H(z)
num_z , den_z = sig.bilinear(num_s,den_s, fs = fs)

# Obtengo H(z)
w_z , h_z = sig.freqz(num_z, den_z)

#filtro_digital = sig.TransferFunction(num_z , den_z , dt = 1/fs)
#analyze_sys([filtro_digital],['prueba'])

# Ploteo
plt.close('all')
fig,ax = plt.subplots(2,1)

ax[0].set_title('Respuesta de módulo')
ax[0].semilogx(w_z*fs, 20 * np.log10(abs(h_z)), 'b--', w_s, 20 * np.log10(abs(h_s)), 'g')
ax[0].set_ylabel('Amplitud [dB]')
ax[0].set_xlabel('Frecuencia [rad/seg]')
ax[0].grid(True)
ax[0].legend(['H(z)', 'H(s)'])
ax[0].set_xlim([1, 1e6])
ax[0].set_ylim([-160, 10])

angles_z = np.unwrap(np.angle(h_z))
angles_s = np.unwrap(np.angle(h_s))

ax[1].set_title('Respuesta de fase')
ax[1].semilogx(w_z*fs, angles_z, 'b--', w_s, angles_s, 'g')
ax[1].set_ylabel('Angulo (radianes)')
ax[1].grid(True)
ax[1].legend(['H(z)', 'H(s)'])
ax[1].set_xlim([1, 1e6])
ax[1].set_ylim([-3.5, 0])

fig.tight_layout()
```

#### d) Hace falta usar prewarping?

En el punto a), en el cual tenemos un filtro pasa bajos con frecuencia de corte de 1KHz (wc = 6280 rad/seg) y una fs = 100KHz , no haría falta usar prewarping, ya que como vemos en los gráficos la distorsión de la respuesta del filtro digital comienza en aproximadamente 100.000 rad/seg.

Sin embargo, en el punto b), al disminuir la frecuencia de sampleo a 100KHz, la distorsión comienza antes, aproximadamente en 10.000 rad/seg. Este valor es muy cercano a la frecuencia de corte del filtro, por lo que habría que usar prewarping.

Por último, en el punto c), al ser un filtro pasa altos voy a tener que usar prewarping ya que sino la banda de paso queda afectada.
