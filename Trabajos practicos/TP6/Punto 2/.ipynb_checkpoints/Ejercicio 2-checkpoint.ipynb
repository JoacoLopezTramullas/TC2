### a) Cauer 2 (remociÃ³n en infinito)

```python
import sympy as sp
import splane as tc2
from schemdraw import Drawing
from schemdraw.elements import  Resistor, Capacitor, Inductor
import numpy as np

s = sp.symbols('s ', complex=True)

# Sea la siguiente funciÃ³n de excitaciÃ³n
Y_s = (s**5+ 6*s**3 + 8*s)/(s**4 +4*s**2 +3)

# Aplico remociones en infinito
Y2, Y_rem_1 = tc2.remover_polo_infinito(Y_s)
Z4, Z_rem_3 = tc2.remover_polo_infinito(1/Y2)
Y6, Y_rem_5 = tc2.remover_polo_infinito(1/Z4)
Z8, Z_rem_7 = tc2.remover_polo_infinito(1/Y6)
Y10, Y_rem_9 = tc2.remover_polo_infinito(1/Z8)

# Calculo componentes
C1 = Y_rem_1/s
L1 = Z_rem_3/s
C2 = Y_rem_5/s
L2 = Z_rem_7/s
C3 = Y_rem_9/s

# Dibujamos la red resultante
d = Drawing(unit=4)

d = tc2.dibujar_puerto_entrada(d,
                        voltage_lbl = ('+', '$V$', '-'), 
                        current_lbl = '$I$')

d, zz_lbl = tc2.dibujar_funcion_exc_abajo(d, 
                                          'Y',  
                                          Y_s, 
                                          hacia_salida = True,
                                          k_gap_width = 0.5)

d = tc2.dibujar_elemento_derivacion(d, Capacitor, C1)

d = tc2.dibujar_espacio_derivacion(d)

d = tc2.dibujar_elemento_serie(d, Inductor, L1)
    
d = tc2.dibujar_elemento_derivacion(d, Capacitor, C2)

d = tc2.dibujar_elemento_serie(d, Inductor, L2)

d = tc2.dibujar_elemento_derivacion(d, Capacitor, C3)

display(d)
```
