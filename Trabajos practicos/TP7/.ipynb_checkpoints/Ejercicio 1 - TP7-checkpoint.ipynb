## Ejercicio 1A

```python
import ltspice
import matplotlib.pyplot as plt
import scipy.signal as sig
import sympy as sim
import numpy as np

l = ltspice.Ltspice("./Simulaciones/EJ_1A.raw")
l.parse()

# Obtengo curvas de simulacin en ltspice
freq = l.get_frequency()
Y21 =l.get_data('Y21(V1)')
Y22 =l.get_data('Y22(V1)')

# Calculo amplitud y fase de Y11 e Y21
Y21_amplitude = 20 * np.log10(np.abs(Y21))
Y22_amplitude = 20 * np.log10(np.abs(Y22))

# Defino num y den de Y11 e Y21 
Y21_num = [1, 1, 0]
Y21_den = [1 ,2]

Y22_num = [1, 2.25, 0]
Y22_den = [1, 2]

# Obtengo Y11(s) e Y21(s)
w, Y21_s = sig.freqs(Y21_num,Y21_den,freq*2*np.pi)
w, Y22_s = sig.freqs(Y22_num,Y22_den,freq*2*np.pi)

# Ploteo
plt.close('all')
fig, axes = plt.subplots(2, 1, figsize=(15,10))

axes[0].semilogx(freq*2*np.pi, Y21_amplitude, linewidth=3.0, label='LTspice')
axes[0].semilogx(w, 20*np.log10(np.abs(Y21_s)), 'r--', linewidth=3.0, label='Analtico')
axes[0].set_title("|Y21|")
axes[0].set_ylabel("Amplitud")
axes[0].set_xlim([1,10000])
axes[0].legend()
axes[0].grid()

axes[1].semilogx(freq*2*np.pi, Y22_amplitude, linewidth=3.0, label='LTspice')
axes[1].semilogx(w, 20*np.log10(np.abs(Y22_s)), 'r--', linewidth=3.0, label='Analtico')
axes[1].set_title("Mdulo de Y22")
axes[1].set_ylabel("Amplitud")
axes[1].set_xlim([1,10000])
axes[1].legend()
axes[1].grid()
```

## Ejercicio 1b

```python
import ltspice
import matplotlib.pyplot as plt
import scipy.signal as sig
import sympy as sim
import numpy as np

l = ltspice.Ltspice("./Simulaciones/EJ_1B.raw")
l.parse()

# Obtengo curvas de simulacin en ltspice
freq = l.get_frequency()
Z21 =l.get_data('Z21(V1)')
Z22 =l.get_data('Z22(V1)')

# Escalo bien la Z21 ya que analiticamente da con una constante
k = 1.5
Z21 = Z21 / k

# Calculo amplitud y fase de Z21 e Z22
Z21_amplitude = 20 * np.log10(np.abs(Z21))
Z22_amplitude = 20 * np.log10(np.abs(Z22))

# Defino num y den de Z21 e Z22
Z21_num = [1, 0]
Z21_den = [1, 0, 2]

Z22_num = [2, 0, 1]
Z22_den = [1, 0, 2, 0]

# Obtengo Z21(s) e Z22(s)
w, Z21_s = sig.freqs(Z21_num,Z21_den,freq*2*np.pi)
w, Z22_s = sig.freqs(Z22_num,Z22_den,freq*2*np.pi)

# Ploteo
plt.close('all')
fig, axes = plt.subplots(2, 1, figsize=(15,10))

axes[0].semilogx(freq*2*np.pi, Z21_amplitude, linewidth=3.0, label='LTspice')
axes[0].semilogx(w, 20*np.log10(np.abs(Z21_s)), 'r--', linewidth=3.0, label='Analtico')
axes[0].set_title("|Z21|")
axes[0].set_ylabel("Amplitud")
axes[0].set_xlim([0.01,1000])
axes[0].legend()
axes[0].grid()

axes[1].semilogx(freq*2*np.pi, Z22_amplitude, linewidth=3.0, label='LTspice')
axes[1].semilogx(w,20*np.log10(np.abs(Z22_s)), 'r--', linewidth=3.0, label='Analtico')
axes[1].set_title("|Z22|")
axes[1].set_ylabel("Amplitud")
axes[1].set_xlim([0.01,1000])
axes[1].legend()
axes[1].grid()
```
