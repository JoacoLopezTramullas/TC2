## Ejercicio 1

```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt

# módulo de SciPy
from scipy import signal as sig

# Agrego funciones que voy a usar de splane
from splane import bodePlot, pzmap

# Ajusto tamaño de graficos
mpl.rcParams['figure.figsize'] = (10,10)

a = 0.131652809

# Defino numerador y denominador
num = np.array([1,-a])
den = np.array([1,a])

# Obtengo transferencia
H_s = sig.TransferFunction(num,den)

# Ploteo
plt.close('all')
bodePlot(H_s, [-3,6], [-3,3])

# Diagrama de polos y ceros
pzmap(H_s)
```

#### Circuito pasivo:





Vemos que en w=1 rad/seg (159mHz) tenemos una rotación de fase de 15º, ya que la fase paso de valer 180º a valer 165º.

#### Circuito activo:





Vemos que en w=1 rad/seg (159mHz) tenemos una rotación de fase de 15º, ya que la fase paso de valer 0º a valer -15º.
