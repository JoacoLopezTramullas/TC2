```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt

# módulo de SciPy
from scipy import signal as sig

# Agrego funciones que voy a usar de splane
from splane import bodePlot, pzmap

# Ajusto tamaño de graficos
mpl.rcParams['figure.figsize'] = (10,10)
```

#### Transferencia Pasa - Bajos normalizada

```python
# Defino numerador y denominador
num = np.array([1/9,0,1])
den = np.array([1,2,2,1])

# Obtengo transferencia
H_s = sig.TransferFunction(num,den)

# Ploteo
plt.close('all')
bodePlot(H_s, [-2,2], [5,-80])

# Diagrama de polos y ceros
pzmap(H_s)
```

#### Transferencia Pasa - Altos normalizada

```python
# Defino numerador y denominador
num = np.array([1,0,1/9,0])
den = np.array([1,2,2,1])

# Obtengo transferencia
H_s = sig.TransferFunction(num,den)

# Ploteo
plt.close('all')
bodePlot(H_s, [-2,2], [5,-80])

# Diagrama de polos y ceros
pzmap(H_s)
```
