De forma analítica llegamos a la siguiente transferencia:

$\large T4(s) = \frac{s^2}{s^2 + s.2.cos(\frac{\pi}{8})+1} . \frac{s^2}{s^2 + s.2.cos(\frac{3\pi}{8})+1}  $

Desarrollando obtenemos:

$\large T4(s) = \frac{s^4}{s^4 + s^3.2.[cos(\frac{\pi}{8}) + cos(\frac{3\pi}{8})] + s^2.[2+4.cos(\frac{\pi}{8}).cos(\frac{3\pi}{8})]+ s.2.[cos(\frac{\pi}{8}) + cos(\frac{3\pi}{8})] + 1 }  $

```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt

# módulo de SciPy
from scipy import signal as sig

# Agrego funciones que voy a usar de splane
from splane import bodePlot, pzmap ,pretty_print_bicuad_omegayq

# Ajusto tamaño de graficos
mpl.rcParams['figure.figsize'] = (10,10)

# Defino numerador y denominador
A = np.cos(np.pi/8)
B = np.cos(3*np.pi/8)
num = np.array([1,0,0,0,0])
den = np.array([1,2*(A+B),2+4*A*B,2*(A+B),1])

# Obtengo transferencia
T_s = sig.TransferFunction(num,den)

# Ploteo
plt.close('all')
bodePlot(T_s, [-2,3], [40,-140])

# Diagrama de polos y ceros
pzmap(T_s)
```
