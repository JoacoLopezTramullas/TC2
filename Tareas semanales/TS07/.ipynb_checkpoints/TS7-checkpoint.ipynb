$\large Z11 = \frac{a^2}{y_1 . a^2 + y_2 . a^2 + 2. y_2 . a + y_2 +y_3}  $

$\large Z22 = \frac{1}{y_1 . a^2 + y_2 . a^2 + 2. y_2 . a + y_2 +y_3}  $


$\large Z12 = Z21 = \frac{-a}{y_1 . a^2 + y_2 . a^2 + 2. y_2 . a + y_2 +y_3}  $

Proponemos una relación de transformación 2:1 , por lo que a=2, y valores impedancia 1,2 y 3 de modo que:

$\large Z11 = \frac{2^2}{\frac{1}{1} . 2^2 + \frac{1}{2} . 2^2 + 2. \frac{1}{2} . 2 + \frac{1}{2} +\frac{1}{3}} = 0.45283  $

$\large Z22 = \frac{1}{\frac{1}{1} . 2^2 + \frac{1}{2} . 2^2 + 2. \frac{1}{2} . 2 + \frac{1}{2} +\frac{1}{3}} = 0.1132  $

$\large Z12 = Z21  = \frac{-2}{\frac{1}{1} . 2^2 + \frac{1}{2} . 2^2 + 2. \frac{1}{2} . 2 + \frac{1}{2} +\frac{1}{3}} = -0.2264  $
