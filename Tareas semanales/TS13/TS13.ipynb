## Ejercicio 1



### Simulación simbólica

```python
import sympy as sp
import splane as tc2
from schemdraw import Drawing
from schemdraw.elements import  Resistor, Capacitor, Inductor

s = sp.symbols('s ', complex=True)

Z22 = (s**2+8*s+12)/(s**2+5*s+4)

# Remuevo parcialmente valor constante en infinito a Z222
Z4, Zrem_2 = tc2.remover_valor(Z22, 2.298437881)

# Remuevo polo en sigma = -1
sigma = 1
Z6, Zrem_4, R2, C2 = tc2.remover_polo_sigma(Z4, sigma, isImpedance = True, isRC = True, sigma_zero = None)

# Remuevo polo en sigma = -4
sigma = 4
Z8, Zrem_6, R3, C3 = tc2.remover_polo_sigma(Z6, sigma, isImpedance = True, isRC = True, sigma_zero = None)

# Remuevo valor en infinito 
Z10, Zrem_8 = tc2.remover_valor_en_infinito(Z8)

# Calculo componentes
R1 = Zrem_2
R4 = Zrem_8

# Dibujamos la red resultante
d = Drawing(unit=4)

d = tc2.dibujar_puerto_entrada(d,
                        voltage_lbl = ('+', '$V$', '-'), 
                        current_lbl = '$I$')

d, zz_lbl = tc2.dibujar_funcion_exc_abajo(d, 
                                          'Z22',  
                                          Z22, 
                                          hacia_salida = False,
                                          k_gap_width = 0.5)

d = tc2.dibujar_elemento_derivacion(d, Resistor, R4)
d = tc2.dibujar_espacio_derivacion(d)
d = tc2.dibujar_tanque_RC_serie(d, R3, C3)
d = tc2.dibujar_espacio_derivacion(d)
d = tc2.dibujar_tanque_RC_serie(d, R2, C2)
d = tc2.dibujar_espacio_derivacion(d)
d = tc2.dibujar_elemento_serie(d, Resistor, R1)
display(d)
```

### Simulación circuital



```python
import ltspice
import matplotlib.pyplot as plt
import scipy.signal as sig
import sympy as sim
import numpy as np

l = ltspice.Ltspice("./Simulaciones/Ej1.raw")
l.parse()

# Obtengo curvas de simulación en ltspice
freq = l.get_frequency()
T =l.get_data('I(RL)')

# Calculo |T(s)|
T_amplitude = 20 * np.log10(np.abs(T))

# Defino num y den de T(s)
T_num = [1/12, 5/12, 4/12]
T_den = [1, 8, 12]

# Obtengo T(s)
w, T_s = sig.freqs(T_num,T_den,freq*2*np.pi)

# Ploteo
plt.close('all')
fig, axes = plt.subplots(1, 1, figsize=(15,10))

axes.semilogx(freq*2*np.pi, T_amplitude, linewidth=3.0, label='LTspice')
axes.semilogx(w, 20*np.log10(np.abs(T_s)), 'r--', linewidth=3.0, label='Analítico')
axes.set_title("|T(s)|")
axes.set_ylabel("Amplitude")
axes.set_xlim([0.01,100])
axes.set_ylim([-32,-20])
axes.legend()
axes.grid()
```

```python
import sympy as sp
import splane as tc2
from schemdraw import Drawing
from schemdraw.elements import  Resistor, Capacitor, Inductor

s = sp.symbols('s ', complex=True)

Z21 = (s**2+9)/(s**3+2*s)
Z22 = (2*s**2+1)/(s**3+2*s)

Y22 = 1/Z22

# Aplico remociones

# Remuevo parcialmente polo en infinito para logar cero en w = 3
Y4, Yrem_2 = tc2.remover_polo_infinito(Y22, omega_zero = 3 )

Z4 = 1/Y4

# Remuevo polo en w = 3
omega_L2C2 = 3
Z6 , Zrem_4, L2, C2 = tc2.remover_polo_jw(Z4,omega = omega_L2C2, isImpedance = True) 

Y6 = 1/Z6

# Calculo componentes
C3 = Yrem_2/s
C1 = Y6/s
    
# Dibujamos la red resultante
d = Drawing(unit=4)

d = tc2.dibujar_puerto_entrada(d,
                        voltage_lbl = ('+', '$V$', '-'), 
                        current_lbl = '$I$')

d, zz_lbl = tc2.dibujar_funcion_exc_abajo(d, 
                                          'Z22',  
                                          Z22, 
                                          hacia_salida = False,
                                          k_gap_width = 0.5)

d = tc2.dibujar_elemento_derivacion(d, Capacitor, C1)
d = tc2.dibujar_tanque_serie(d, L2, C2)
d = tc2.dibujar_elemento_derivacion(d, Capacitor, C3)
display(d)
```

### Simulación circuital

```python
import ltspice
import matplotlib.pyplot as plt
import scipy.signal as sig
import sympy as sim
import numpy as np

l = ltspice.Ltspice("./Simulaciones/Ej2.raw")
l.parse()

# Obtengo curvas de simulación en ltspice
freq = l.get_frequency()
T =l.get_data('V(V2)')

# Calculo |T(s)|
T_amplitude = 20 * np.log10(np.abs(T))

# Defino num y den de T(s)
T_num = [1, 0, 9]
T_den = [1, 2, 2, 1]

# Obtengo T(s)
w, T_s = sig.freqs(T_num,T_den,freq*2*np.pi)

k = 1/9
T_s = k * T_s

# Ploteo
plt.close('all')
fig, axes = plt.subplots(1, 1, figsize=(15,10))

axes.semilogx(freq*2*np.pi, T_amplitude, linewidth=3.0, label='LTspice')
axes.semilogx(w, 20*np.log10(np.abs(T_s)), 'r--', linewidth=3.0, label='Analítico')
axes.set_title("|T(s)|")
axes.set_ylabel("Amplitude")
axes.set_xlim([0.01,100])
axes.set_ylim([-55,5])
axes.legend()
axes.grid()
```
